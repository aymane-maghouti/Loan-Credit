#  Data cleaning

### strategy cleaning :

For categorical (non-numeric) variables, we will fill the missing values with the values that repeat the most.

For numeric variables, we will fill the missing values with the values next to them.

```python
cat_data=[]
num_data=[]
for i,j in enumerate(df.dtypes):
    if j==object :
        cat_data.append(df.iloc[:,i])
    else :
        num_data.append(df.iloc[:,i])
        
```

```python
num_data
```

```python
cat_data
```

Transformation of cat_data and num_data lists into data frames
(we will use the transpose method just for keeping the standard format of a DataFrame)

```python
cat_data = pd.DataFrame(cat_data).transpose()
cat_data
```

```python
num_data = pd.DataFrame(num_data).transpose()
num_data
```

# cleaning numeric data

```python
num_data.fillna(method='bfill',inplace=True)
num_data
```

```python
num_data.isnull().sum().sort_values(ascending=False)
```

### cleaning categorical data

```python
cat_data = cat_data.apply(lambda x:x.fillna(x.value_counts().index[0]))
cat_data
```

```python
cat_data.isnull().sum().sort_values(ascending=False)
```

# Data preparation
Data transformation of the variable target (Loan_Status) -i.e- (Y-->1 , N-->0)

```python
target_val = {'Y':1,'N':0}
target = cat_data['Loan_Status']
cat_data.drop('Loan_Status',axis=1,inplace=True)
target=target.map(target_val)
target
```

Now we will use the labelEncoder methods for the other categorical data with the same logic that we use on target data.

```python
le = LabelEncoder()
for i in cat_data:
    cat_data[i] = le.fit_transform(cat_data[i])
cat_data
```

we can see that the variable Loan_Id is not important in our study so we will delete it !

```python
cat_data.drop('Loan_ID',axis=1,inplace=True)
```

```python
cat_data
```

now it's time to combine transformed categorical data and numerical data 

```python
X = pd.concat([cat_data,num_data],axis=1)
Y = target
X
```

# model building

### dataset Spliting (Train set & Test set)

```python
sss = StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train,test in sss.split(X,Y):
    X_train,X_test = X.iloc[train],X.iloc[test]
    Y_train,Y_test = Y.iloc[train],Y.iloc[test]
```

### machine learning model 
For building the machine learning model we will apply two algorithms, logistic regression and DecisionTree, then we will evaluate each algorithm and we will choose the best

```python
models = {'LogisticRegression':LogisticRegression(random_state=42),
          'DecisionTreeClassifier':DecisionTreeClassifier(max_depth=1,random_state=42)}
```

To evaluate your models we will use the accuracy score metric (we are going to develop a function that will be based on this metric in order to evaluate our models) 

```python
def accuracy(y_true,y_pred,retu=False):
    acc = accuracy_score(y_true,y_pred)
    if retu:
        return acc 
    else :
        print(f'the accuracy of the medel is: {acc*100} %')
def evaluation(models,X_train,X_test,Y_train,Y_test):
    for name,model in models.items():
        print(f'{name} :')
        model.fit(X_train,Y_train)
        accuracy(Y_test,model.predict(X_test))
        print('-'*60)
evaluation(models,X_train,X_test,Y_train,Y_test)
```

We can clearly see that the performances of the two algorithms are almost equal, but we will choose the model based on the Logistic Regression algorithm (≈85.37%)

Now we are going to build a new dataset which contains just the variables which have a great impact on the target variable based on Data exploratory analysis  

```python
X_2=X[['Credit_History','Gender','Married','CoapplicantIncome','ApplicantIncome']]
X_2
```

```python
sss = StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train,test in sss.split(X_2,Y):
    X_train,X_test = X_2.iloc[train],X_2.iloc[test]
    Y_train,Y_test = Y.iloc[train],Y.iloc[test]
print('X train set size :',X_train.shape)
print('X test set size :',X_test.shape)
print('Y train set size :',Y_train.shape[0])
print('Y test set size :',Y_test.shape[0])
```

```python
evaluation(models,X_train,X_test,Y_train,Y_test)
```
